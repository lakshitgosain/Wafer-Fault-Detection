- A wafer is a very thin slice of Semi conducgter material which is used in various electronic devices....a production line, various wafers are deployed....they have to check which  wafer is faulty and replace the wafer

We need to predict either -1 or +1....hence it is a binary classification problem

Data transformation-there are certain things we need to do before inserting values into our DB...we need to explictily convert some values to NULL to insert to DB.....and like removing single inverted comas with double inverted comas etc....

we will then create a single csv file which will contain all the valid data extracted from the batch files.....this will act as input for the subsequesnt steps....On this data , we will perform Data preprocessing and EDA(this is the next step).
